```python
# Read data

fert = pd.read_csv('input_data/gapminder_total_fertility.csv', index_col=0)
life = pd.read_excel('input_data/gapminder_lifeexpectancy.xlsx', index_col=0)
population = pd.read_excel('input_data/gapminder_population.xlsx', index_col=0)
```

```python
# Clean data

fert.columns = fert.columns.astype(int)
fert.index.name = 'country'
life.index.name = 'country'
population.index.name = 'country'
```

```python
# Change format wide -> long

fert = fert.reset_index()
fert = fert.melt(id_vars='country', var_name='year', value_name='fertility_rate')

life = life.reset_index()
life = life.melt(id_vars='country', var_name='year', value_name='life_expectancy')

population = population.reset_index()
population = population.melt(id_vars='country', var_name='year', value_name='population')
```

```python
# Merge data

df = fert.merge(population)
df = df.merge(life)
```

```python
# Plot example

df_subset = df.loc[df['country'].isin(['France', 'Germany', 'Sweden'])]

sns.lineplot(
    x='year',
    y='life_expectancy',
    hue='country',
    data=df_subset
)
plt.figure()

sns.scatterplot(
    x='year', 
    y='life_expectancy',
    hue='country',
    data=df_subset,
    alpha=0.6
)
plt.figure()

sns.scatterplot(
    x='year',
    y='life_expectancy',
    hue='country',
    size='fertility_rate',
    data=df_subset, alpha=0.6
)
plt.figure()
```

```python
# Set countries

df['country'].unique()
countries = [
    'Finland',
    'Germany', 
    'Ghana',
    'Malta',
    'Poland',
    'United Kingdom',
    'Vietnam',
]
```

```python
# Generate plots

for year in range(1960, 2016):
    plt.figure()
    df_subset = df[df['country'].isin(countries)]
    df_subset_year = df_subset[df_subset['year'] == year]
    plot = sns.scatterplot(
        x='population',
        y='life_expectancy',
        hue='country',
        data=df_subset_year,
        alpha=1
    )
    plt.axis((0, 100000000, 20, 100))
    fig = plot.get_figure()
    fig.savefig(f'plots/lifeexp_{year}.png')
    plt.close()
```

```python
# Generate gif

import imageio

images = []

for year in range(1960, 2016):
    images.append(imageio.imread(f'plots/lifeexp_{year}.png'))

imageio.mimsave('rosling.gif', images, fps=10)
```
